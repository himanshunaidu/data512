# Rare Disease Dataset: Monthly Article Traffic

This notebook contains code to construct, analyze, and publish a dataset of monthly article traffic for a select set of rare diseases from English Wikipedia from July 1, 2015 through September 30, 2024.

This list of diseases and their relevant Wikipedia pages was collected by using a database of rare diseases maintained by the [National Organization for Rare Diseases (NORD)](https://rarediseases.org/) and matching them to Wikipedia articles that are either about a rare disease or have a section that mentions a rare disease.


## Access Page View Data

The following sub-section contains code to access page view data using the [Wikimedia REST API](https://www.mediawiki.org/wiki/Wikimedia_REST_API). The API documentation, [pageviews/per-article](https://wikimedia.org/api/rest_v1/#/Pageviews%20data), covers additional details that may be helpful when trying to use or understand this example.

```python
#
# These are standard python modules
import json, time, urllib.parse
#
# The 'requests' module is not a standard Python module. You will need to install this with pip/pip3 if you do not already have it
import requests
```

The example relies on some constants that help make the code a bit more readable.

```python
#########
#
#    CONSTANTS
#

# The REST API 'pageviews' URL - this is the common URL/endpoint for all 'pageviews' API requests
API_REQUEST_PAGEVIEWS_ENDPOINT = 'https://wikimedia.org/api/rest_v1/metrics/pageviews/'

# This is a parameterized string that specifies what kind of pageviews request we are going to make
# In this case it will be a 'per-article' based request. The string is a format string so that we can
# replace each parameter with an appropriate value before making the request
API_REQUEST_PER_ARTICLE_PARAMS = 'per-article/{project}/{access}/{agent}/{article}/{granularity}/{start}/{end}'

# The Pageviews API asks that we not exceed 100 requests per second, we add a small delay to each request
API_LATENCY_ASSUMED = 0.002       # Assuming roughly 2ms latency on the API and network
API_THROTTLE_WAIT = (1.0/100.0)-API_LATENCY_ASSUMED

# When making a request to the Wikimedia API they ask that you include your email address which will allow them
# to contact you if something happens - such as - your code exceeding rate limits - or some other error
REQUEST_HEADERS = {
    'User-Agent': 'hnaidu36@uw.edu, University of Washington, MSDS DATA 512 - AUTUMN 2023',
}

# This is just a list of English Wikipedia article titles that we can use for example requests
ARTICLE_TITLES = [ 'Bison', 'Northern flicker', 'Red squirrel', 'Chinook salmon', 'Horseshoe bat' ]

# This template is used to map parameter values into the API_REQUST_PER_ARTICLE_PARAMS portion of an API request. The dictionary has a
# field/key for each of the required parameters. In the example, below, we only vary the article name, so the majority of the fields
# can stay constant for each request. Of course, these values *could* be changed if necessary.
ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE = {
    "project":     "en.wikipedia.org",
    "access":      "desktop",      # this should be changed for the different access types
    "agent":       "user",
    "article":     "",             # this value will be set/changed before each request
    "granularity": "monthly",
    "start":       "2015010100",   # start and end dates need to be set
    "end":         "2023040100"    # this is likely the wrong end date
}
```

The API request will be made using one procedure. The idea is to make this reusable. The procedure is parameterized, but relies on the constants above for the important parameters. The underlying assumption is that this will be used to request data for a set of article pages. Therefore the parameter most likely to change is the article_title.

```python
#########
#
#    PROCEDURES/FUNCTIONS
#

def request_pageviews_per_article(article_title = None,
                                  endpoint_url = API_REQUEST_PAGEVIEWS_ENDPOINT,
                                  endpoint_params = API_REQUEST_PER_ARTICLE_PARAMS,
                                  request_template = ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE,
                                  headers = REQUEST_HEADERS):

    # article title can be as a parameter to the call or in the request_template
    if article_title:
        request_template['article'] = article_title

    if not request_template['article']:
        raise Exception("Must supply an article title to make a pageviews request.")

    # Titles are supposed to have spaces replaced with "_" and be URL encoded
    article_title_encoded = urllib.parse.quote(request_template['article'].replace(' ','_'), safe='')
    request_template['article'] = article_title_encoded

    # now, create a request URL by combining the endpoint_url with the parameters for the request
    request_url = endpoint_url+endpoint_params.format(**request_template)

    # make the request
    try:
        # we'll wait first, to make sure we don't exceed the limit in the situation where an exception
        # occurs during the request processing - throttling is always a good practice with a free
        # data source like Wikipedia - or other community sources
        if API_THROTTLE_WAIT > 0.0:
            time.sleep(API_THROTTLE_WAIT)
        response = requests.get(request_url, headers=headers)
        json_response = response.json()
    except Exception as e:
        print(e)
        json_response = None
    return json_response
```

## Rare Disease Data Acquisition

The Data Acquisition step uses the pageviews API to collect data on rare diseases. This data includes counts of user pageviews, for different access types (desktop, mobile web and mobile app).

This step produces three files:
1. Monthly Mobile Access:

Filename: [rare-disease_monthly_mobile_201501-202409.json](data/rare-disease_monthly_mobile_201501-202409.json)

Description: This file contains the sum of mobile web and mobile app traffic for each article for each month in the specified date range.

2. Monthly Desktop Access:

Filename: [rare-disease_monthly_desktop_201501-202409.json](data/rare-disease_monthly_desktop_201501-202409.json)

Description: This file contains the desktop traffic for each article for each month in the specified date range.

3. Monthly Cumulative Access:

Filename: [rare-disease_monthly_cumulative_201501-202409.json](data/rare-disease_monthly_cumulative_201501-202409.json)

Description: This file contains the sum of all mobile and desktop traffic for each article for each month in the specified date range.

IMPORTS

```python
import os

# The following packages are not standard Python modules. You will need to install this with pip/pip3/conda if you do not already have it
import numpy as np
import pandas as pd

from tqdm import tqdm
```

CONSTANTS

```python
# Dev environment flag to control the number of requests made to the API for testing
DEV_ENVIRONMENT = False

# The cleaned CSV file path that contains the rare disease names
RARE_DISEASE_CSV_PATH = "rare-disease_cleaned.AUG.2024.csv"

# The start and end dates of the data we want to collect from the Pageviews API
START = "2015010100"
END = "2024093000"
```

```python
# The start and end dates trimmed to only include the year and month
# This is used to name the output files
START_DATE = START[0:-4]
END_DATE = END[0:-4]

# The path to the data directory
DATA_PATH = "data"
# The file names of the article views data extracted from the Pageviews API
DESKTOP_VIEWS_FILE = f"rare-disease_monthly_desktop_{START_DATE}-{END_DATE}.json"
MOBILE_VIEWS_FILE = f"rare-disease_monthly_mobile_{START_DATE}-{END_DATE}.json"
CUMULATIVE_VIEWS_FILE = f"rare-disease_monthly_cumulative_{START_DATE}-{END_DATE}.json"
```

```python
# Change the arguments of params templates
ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE['start'] = START
ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE['end'] = END

# Create access-specific params templates
ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE_DESKTOP = ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE.copy()
ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE_DESKTOP['access'] = "desktop"

ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE_MOBILE_WEB = ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE.copy()
ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE_MOBILE_WEB['access'] = "mobile-web"

ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE_MOBILE_APP = ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE.copy()
ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE_MOBILE_APP['access'] = "mobile-app"
```

FUNCTIONS

```python
def write_json_to_file(data, filename, mode = "a+"):
    with open(filename, mode) as json_file:
        json.dump(data, json_file, ensure_ascii=False)
```

```python
def get_page_views_df_for_article(article_title = None,
                                  endpoint_url = API_REQUEST_PAGEVIEWS_ENDPOINT,
                                  request_template = ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE,
                                  endpoint_params = API_REQUEST_PER_ARTICLE_PARAMS,
                                  headers = REQUEST_HEADERS):
    '''
    Get page views for a given article title.
    This drops the 'access' key from each of the response items.

    Parameters:
    ----------
    article_title: str
        The title of the article to get page views for
    
    endpoint_url: str
        The URL of the endpoint to make the request to

    request_template: dict
        The template for the request parameters
    
    endpoint_params: str
        The parameters for the request.
        These will be used to format the request_template

    headers: dict
        The headers to be used in the request
    
    Returns:
    -------
    response: pd.DataFrame
        The response from the API
        
    '''
    views = request_pageviews_per_article(article_title, endpoint_url, endpoint_params, request_template, headers)
    views = views['items']
    views_df = pd.DataFrame(views)
    views_df.drop(columns = ['access'], inplace = True)
    return views_df
```

### Load CSV

```python
# Read the rare disease names from the cleaned CSV file
rare_disease_df = pd.read_csv(os.path.join(os.getcwd(), RARE_DISEASE_CSV_PATH))

# The following dictionaries will store the page views data for each article
views_desktop = {}
views_mobile = {}
views_cumulative = {}
```

Sort the dataframe by disease names so that the resulting json files are also ordered by disease names

```python
rare_disease_df.sort_values(by='disease', inplace=True)
```

```python
if DEV_ENVIRONMENT:
    rare_disease_df = rare_disease_df.head(20)
```

### Query API

The following loop will make API requests for each article in the list of article titles. The request returns a dataframe with the page views for the article. The dataframe is then stored as a JSON array in a dictionary with the article title as the key.

Note: The following loop was executed on a system equipped with a 12th Gen Intel® Core™ i7-12700H processor (2.30 GHz). The total execution time for this loop was 31 minutes and 40 seconds.

```python
# The following columns will be used to merge the page views dataframes in order to sum the relevant views
MERGE_COLUMNS = ["project", "article", "granularity","timestamp", "agent"]

for index, row in tqdm(rare_disease_df.iterrows(), total=rare_disease_df.shape[0]):
    article = row['disease']
    
    # Get desktop views
    views_desktop_df = get_page_views_df_for_article(article, request_template=ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE_DESKTOP)
    views_desktop[article] = json.loads(views_desktop_df.to_json(orient = "records"))

    # Get mobile views
    views_mobile_web_df = get_page_views_df_for_article(article, request_template=ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE_MOBILE_WEB)
    views_mobile_app_df = get_page_views_df_for_article(article, request_template=ARTICLE_PAGEVIEWS_PARAMS_TEMPLATE_MOBILE_APP)

    views_mobile_df = pd.merge(views_mobile_web_df, views_mobile_app_df, on=MERGE_COLUMNS, how = "outer", suffixes=('_df1', '_df2'))
    views_mobile_df['views'] = views_mobile_df['views_df1'].fillna(0) + views_mobile_df['views_df2'].fillna(0)
    views_mobile_df.drop(columns = ['views_df1', 'views_df2'], inplace = True)

    views_mobile[article] = json.loads(views_mobile_df.to_json(orient = "records"))

    # Get cumulative views
    views_cumulative_df = pd.merge(views_desktop_df, views_mobile_df, on=MERGE_COLUMNS, how = "outer", suffixes=('_df1', '_df2'))
    views_cumulative_df['views'] = views_cumulative_df['views_df1'].fillna(0) + views_cumulative_df['views_df2'].fillna(0)
    views_cumulative_df.drop(columns = ['views_df1', 'views_df2'], inplace = True)

    views_cumulative[article] = json.loads(views_cumulative_df.to_json(orient = "records"))

    if index+1 % 100 == 0:
        tqdm.write(f"Processed {index} articles")
```

### Write Views to Files

```python
# Write the page views data to JSON files
write_json_to_file(views_desktop, os.path.join(DATA_PATH, DESKTOP_VIEWS_FILE), "w")
write_json_to_file(views_mobile, os.path.join(DATA_PATH, MOBILE_VIEWS_FILE), "w")
write_json_to_file(views_cumulative, os.path.join(DATA_PATH, CUMULATIVE_VIEWS_FILE), "w")
```

## Analysis

Set up the relevant imports.

Add the relevant constant for saving the plots.

Create the necessary functions that will get the necessary aggregate information from view data, retrieve the articles of interest depending on the analysis, and extract time series data from the page view files generated using the pageviews API.

IMPORTS

```python
import matplotlib.pyplot as plt
```

CONSTANTS

```python
# The path to the plots directory. This is where the plots will be saved
PLOT_PATH = "plots"
```

FUNCTIONS

```python
# Get json data from a file
def get_json_from_file(filename):
    with open(filename, "r") as json_file:
        data = json.load(json_file)
    return data
```

```python
def get_aggregates_for_article(views_df, article_title):
    '''
    Get the maximum, minimum, average and other values from views for an article

    Parameters:
    ----------
    views_df: pd.DataFrame
        The dataframe containing the views data for an article
    
    article_title: str
        The title of the article
    
    Returns:
    -------
    aggreate_views: dict
        A dictionary containing the maximum, minimum and average views for an article
    '''
    max_views = views_df['views'].max()
    max_month = views_df.loc[views_df['views'].idxmax()]['timestamp']
    min_views = views_df['views'].min()
    min_month = views_df.loc[views_df['views'].idxmin()]['timestamp']
    avg_views = views_df['views'].mean()
    count_views = views_df['views'].count()

    if article_title is None and views_df.shape[0] > 0:
        article_title = views_df['article'].iloc[0]

    aggregate_views = {
        'article': article_title,
        'max_views': max_views,
        'max_month': max_month,
        'min_views': min_views,
        'min_month': min_month,
        'avg_views': avg_views,
        'count_views': count_views
    }

    return aggregate_views
```

```python
def get_aggregate_df_from_views_json(views_json):
    '''
    Get the aggregate dataframe from views json data.

    The json file is expected to have article titles as keys and views data as values.

    Parameters:
    ----------
    filename: str
        The name of the json file to read from

    Returns:
    -------
    pd.DataFrame
        The aggregate dataframe from the json file
    '''
    aggregate_array = []

    for article_title, views_df in views_json.items():
        views_df = pd.DataFrame(views_df)
        aggregate_views = get_aggregates_for_article(views_df, article_title)
        aggregate_array.append(aggregate_views)
    
    return pd.DataFrame(aggregate_array)
```

```python
# Get the article with the maximum average views
def get_article_with_max_views_average(aggregated_df):
    return aggregated_df.iloc[aggregated_df["avg_views"].idxmax()]

# Get the article with the minimum average views
def get_article_with_min_views_average(aggregated_df):
    return aggregated_df.iloc[aggregated_df["avg_views"].idxmin()]

# Get the top 10 articles with the maximum peak views
def get_top_peak_articles(aggregated_df, count = 10):
    return aggregated_df.nlargest(count, 'max_views')

# Get the top 10 articles with the fewest months of view data
def get_lowest_count_articles(aggregated_df, count = 10):
    return aggregated_df.nsmallest(count, 'count_views')
```

```python
def get_time_series_for_article_from_view_data(article_title, views):
    '''
    Get the page views data for an article from the views data.
    This page views data is the time series of monthly activity for the article.
    Also, add an extra label to the views data.

    Parameters:
    ----------
    article_title: str
        The title of the article to get the dataframe for

    views: dict
        The dictionary containing the views data

    Returns:
    -------
    pd.DataFrame
        The views dataframe for the article
    '''
    views_df = pd.DataFrame(views[article_title])
    views_df['date'] = pd.to_datetime(views_df['timestamp'], format='%Y%m%d%H')
    return views_df
```

### Get Aggregate Dataframes Per Access Type

```python
# From the json files containing the page views data, get the json data.
# We overwrite the original views variables with the data from the json files
views_desktop = get_json_from_file(os.path.join(DATA_PATH, DESKTOP_VIEWS_FILE))
views_mobile = get_json_from_file(os.path.join(DATA_PATH, MOBILE_VIEWS_FILE))
views_cumulative = get_json_from_file(os.path.join(DATA_PATH, CUMULATIVE_VIEWS_FILE))
```

```python
# From the page views json data, create the aggregate dataframes.
# These dataframes will contain summary statistics such as the maximum, minimum and average views for each article.
aggregate_desktop_df = get_aggregate_df_from_views_json(views_desktop)
aggregate_mobile_df = get_aggregate_df_from_views_json(views_mobile)
aggregate_cumulative_df = get_aggregate_df_from_views_json(views_cumulative)
```

### Analysis Part 1

Plot a graph that contains time series for the articles that have the highest average page requests and the lowest average page requests for desktop access and mobile access over the entire time series.

```python
# Get the articles with the maximum and minimum average views for both desktop and mobile

article_max_views_average_desktop = get_article_with_max_views_average(aggregate_desktop_df)['article']
article_min_views_average_desktop = get_article_with_min_views_average(aggregate_desktop_df)['article']

article_max_views_average_mobile = get_article_with_max_views_average(aggregate_mobile_df)['article']
article_min_views_average_mobile = get_article_with_min_views_average(aggregate_mobile_df)['article']
```

```python
print(article_max_views_average_desktop, article_min_views_average_desktop)
print(article_max_views_average_mobile, article_min_views_average_mobile)
```

```python
# Get the view data (as time series) for the articles with the maximum and minimum average views for both desktop and mobile

article_max_views_desktop_time_series = get_time_series_for_article_from_view_data(article_max_views_average_desktop, views_desktop)
article_min_views_desktop_time_series = get_time_series_for_article_from_view_data(article_min_views_average_desktop, views_desktop)

article_max_views_mobile_time_series = get_time_series_for_article_from_view_data(article_max_views_average_mobile, views_mobile)
article_min_views_mobile_time_series = get_time_series_for_article_from_view_data(article_min_views_average_mobile, views_mobile)
```

```python
# Plot the time series data for the articles with the maximum and minimum average views for both desktop and mobile

plt.figure(figsize=(10, 5))
ax = plt.gca()

article_max_views_desktop_time_series.plot(x='date', y='views', label='Desktop Max: '+article_max_views_average_desktop, ax=ax)
article_min_views_desktop_time_series.plot(x='date', y='views', label='Desktop Min: '+article_min_views_average_desktop, ax=ax)
article_max_views_mobile_time_series.plot(x='date', y='views', label='Mobile Max: '+article_max_views_average_mobile, ax=ax, linestyle='dotted')
article_min_views_mobile_time_series.plot(x='date', y='views', label='Mobile Min: '+article_min_views_average_mobile, ax=ax, linestyle='dotted')

plt.xlabel("Year")
plt.ylabel("Views")
plt.title("Articles with Maximum Average and Minimum Average views for Desktop and Mobile")
plt.legend(bbox_to_anchor=(1, 1))

plt.savefig(os.path.join(PLOT_PATH, "maximum_average_and_minimum_average.png"), bbox_inches='tight')

plt.show()
```

### Analysis Part 2

Plot a graph that contains time series for the top 10 article pages by largest (peak) page views over the entire time series by access type.

```python
# Get the top 10 articles with the maximum peak views for both desktop and mobile

top_10_peak_articles_desktop = get_top_peak_articles(aggregate_desktop_df, 10)['article']

top_10_peak_articles_mobile = get_top_peak_articles(aggregate_mobile_df, 10)['article']
```

```python
# Plot the time series data for the top 10 articles with the maximum peak views

plt.figure(figsize=(10, 5))
ax = plt.gca()

for index, article in top_10_peak_articles_desktop.items():
    peak_article_df = get_time_series_for_article_from_view_data(article, views_desktop)
    peak_article_df.plot(x='date', y='views', label='Desktop: '+article, ax=ax)

for index, article in top_10_peak_articles_mobile.items():
    peak_article_df = get_time_series_for_article_from_view_data(article, views_mobile)
    peak_article_df.plot(x='date', y='views', label='Mobile: '+article, ax=ax, linestyle='dotted')

plt.xlabel("Year")
plt.ylabel("Views")
plt.title("Top 10 Articles by Peak Page Views for Desktop and Mobile")
plt.legend(bbox_to_anchor=(1, 1))

plt.savefig(os.path.join(PLOT_PATH, "top_10_peak_page_views.png"), bbox_inches='tight')

plt.show()
```

### Analysis Part 3

Plot a graph that contains pages that have the fewest months of available data.

```python
# Get the top 10 articles with the fewest months of view data for both desktop and mobile

lowest_10_count_articles_desktop = get_lowest_count_articles(aggregate_desktop_df, 10)['article']

lowest_10_count_articles_mobile = get_lowest_count_articles(aggregate_mobile_df, 10)['article']
```

```python
# Plot the time series data for the top 10 articles with the fewest months of view data

plt.figure(figsize=(10, 5))
ax = plt.gca()

for index, article in lowest_10_count_articles_desktop.items():
    lowest_count_article_df = get_time_series_for_article_from_view_data(article, views_desktop)
    lowest_count_article_df.plot(x='date', y='views', label='Desktop: '+article, ax=ax)

for index, article in lowest_10_count_articles_mobile.items():
    lowest_count_article_df = get_time_series_for_article_from_view_data(article, views_mobile)
    lowest_count_article_df.plot(x='date', y='views', label='Mobile: '+article, ax=ax, linestyle='dotted')

plt.xlabel("Month and Year")
plt.ylabel("Views") 
plt.title("Lowest 10 Articles by Count of Months of Page Views for Desktop and Mobile")
plt.legend(bbox_to_anchor=(1, 1))

plt.savefig(os.path.join(PLOT_PATH, "fewest_months_of_data.png"), bbox_inches='tight')

plt.show()
```
